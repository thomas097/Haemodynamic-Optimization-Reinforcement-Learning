## Action Distribution

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### AmsterdamUMCdb

```python
aumc_train_df = pd.read_csv('../preprocessing/datasets/amsterdam-umc-db/aggregated_full_cohort_1h/train.csv')
aumc_valid_df = pd.read_csv('../preprocessing/datasets/amsterdam-umc-db/aggregated_full_cohort_1h/valid.csv')
aumc_test_df = pd.read_csv('../preprocessing/datasets/amsterdam-umc-db/aggregated_full_cohort_1h/test.csv')
```

```python
def actions_to_table(df):
    action_counts = df.action.value_counts()
    table = np.zeros(25)
    for a, c in action_counts.iteritems():
        table[int(a)] = c
    return table.reshape(5, 5)
    
    
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(17, 5))
ax1.set_title('train')
sns.heatmap(actions_to_table(aumc_train_df), annot=True, fmt='g', ax=ax1, cbar=False)
ax1.set_xlabel('VP dose')
ax1.set_ylabel('IV dose')
ax1.invert_yaxis()

ax2.set_title('valid')
sns.heatmap(actions_to_table(aumc_valid_df), annot=True, fmt='g', ax=ax2, cbar=False)
ax2.set_xlabel('VP dose')
ax2.invert_yaxis()

ax3.set_title('test')
sns.heatmap(actions_to_table(aumc_test_df), annot=True, fmt='g', ax=ax3, cbar=False)
ax3.set_xlabel('VP dose')
ax3.invert_yaxis()

plt.show()
```
